```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
print("OpenCV version:", cv2.__version__)
```

# Cơ bản

### đọc file

```python
img1 = cv2.imread('../data/raw/Apple/Fresh/apple_fresh_001.jpg (1).jpg')
img2 = cv2.imread('../data/raw/Apple/Fresh/apple_fresh_001.jpg (93).jpg')
```

### hiển thị

```python
wight = img1.shape[1]
height = img1.shape[0]
img2 = cv2.resize(img2, (wight, height))
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img1)
ax[0].set_title('Image 1')
ax[1].imshow(img2)
ax[1].set_title('Image 2')
fig.set_tight_layout(True)
plt.show()
```

### chuyển đổi không gian màu

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img1_HSV = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
img1_LAB = cv2.cvtColor(img1, cv2.COLOR_BGR2LAB)
img1_YUV = cv2.cvtColor(img1, cv2.COLOR_BGR2YUV)
fig, ax = plt.subplots(3, 2, figsize=(12, 8))
ax[0, 0].imshow(img1_gray, cmap='gray')
ax[0, 0].set_title('Grayscale')
ax[0, 1].imshow(img1_HSV)
ax[0, 1].set_title('HSV Color Space')
ax[1, 0].imshow(img1_LAB)
ax[1, 0].set_title('LAB Color Space')
ax[1, 1].imshow(img1_YUV)
ax[1, 1].set_title('YUV Color Space')
ax[2, 0].imshow(img1)
ax[2, 0].set_title('Original Image')
ax[2, 1].imshow(img2)
ax[2, 1].set_title('Image 2')
fig.set_tight_layout(True)
plt.show()
```

### histogram

```python
fig, ax = plt.subplots(6, 2, figsize=(8, 24))
hist = cv2.calcHist([img1], [0], None, [256], [0, 256])
hist[255] = 0
ax[0, 0].imshow(img1)
ax[0, 0].set_title('Image 1')
ax[0, 1].bar(range(256), hist[:,0], color='r')
ax[0, 1].set_title('Histogram Image 1 - Channel 0')
hist = cv2.calcHist([img1], [1], None, [256], [0, 256])
hist[255] = 0
ax[1, 0].bar(range(256), hist[:,0], color='g')
ax[1, 0].set_title('Histogram Image 1 - Channel 1')
hist = cv2.calcHist([img1], [2], None, [256], [0, 256])
hist[255] = 0
ax[1, 1].bar(range(256), hist[:,0], color='b')
ax[1, 1].set_title('Histogram Image 1 - Channel 2')

hist = cv2.calcHist([img1_gray], [0], None, [256], [0, 256])
hist[255] = 0
ax[2, 0].imshow(img1_gray, cmap='gray')
ax[2, 0].set_title('Grayscale Image 1')
ax[2, 1].bar(range(256), hist[:,0], color='k')
ax[2, 1].set_title('Histogram Grayscale Image 1')

hist = cv2.calcHist([img1_HSV], [0], None, [256], [0, 256])
hist[0] = 0
ax[3, 0].imshow(img1_HSV)
ax[3, 0].set_title('HSV Image 1')
ax[3, 1].bar(range(256), hist[:,0], color='m')
ax[3, 1].set_title('Histogram HSV Image 1 - Channel 0')

hist = cv2.calcHist([img1_LAB], [0], None, [256], [0, 256])
hist[255] = 0
ax[4, 0].imshow(img1_LAB)
ax[4, 0].set_title('LAB Image 1')
ax[4, 1].bar(range(256), hist[:,0], color='c')
ax[4, 1].set_title('Histogram LAB Image 1 - Channel 0')

hist = cv2.calcHist([img1_YUV], [0], None, [256], [0, 256])
hist[255] = 0
ax[5, 0].imshow(img1_YUV)
ax[5, 0].set_title('YUV Image 1')
ax[5, 1].bar(range(256), hist[:,0], color='y')
ax[5, 1].set_title('Histogram YUV Image 1 - Channel 0')

fig.set_tight_layout(True)


plt.show()
```

```python
img1_HSV_eq = cv2.equalizeHist(img1_HSV[:,:,0])
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
hist = cv2.calcHist([img1_HSV_eq], [0], None, [256], [0, 256])
hist[0] = 0
ax[0].imshow(img1_HSV_eq)
ax[0].set_title('Equalized HSV Channel 0 Image 1')
ax[1].bar(range(256), hist[:,0], color='m')
ax[1].set_title('Histogram Equalized HSV Channel 0 Image 1')
fig.set_tight_layout(True)
plt.show()
```

```python
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
hist = cv2.calcHist([img2_gray], [0], None, [256], [0, 256])
# hist[255] = 0
ax[0, 0].imshow(img2_gray, cmap='gray')
ax[0, 0].set_title('Grayscale Image 2')
ax[0, 1].bar(range(256), hist[:,0], color='k')
ax[0, 1].set_title('Histogram Grayscale Image 2')
img2_gray_eq = cv2.equalizeHist(img2_gray)
hist = cv2.calcHist([img2_gray_eq], [0], None, [256], [0, 256])
ax[1, 0].imshow(img2_gray_eq, cmap='gray')
ax[1, 0].set_title('Equalized Grayscale Image 2')
ax[1, 1].bar(range(256), hist[:,0], color='k')
ax[1, 1].set_title('Histogram Equalized Grayscale Image 2')
fig.set_tight_layout(True)
plt.show()
```
